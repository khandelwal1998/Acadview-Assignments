```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
```

```python
df.head()
```

```python
sns.countplot(df['RainToday'])
```

```python
sns.countplot(df['RainTomorrow'])
```

```python
df.plot.scatter(x='Evaporation',y='Sunshine',c='Rainfall',cmap='coolwarm')
```

```python
sns.countplot(df['WindGustDir'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
df.WindDir9am.value_counts().plot(kind = 'barh')
plt.title("Wind Direction 9 am")
```

```python
df.WindDir3pm.value_counts().plot(kind = 'barh')
plt.title("Wind Direction 3 pm")
```

```python
sns.distplot(df['Rainfall'].dropna())
```
